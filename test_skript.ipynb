```python
import pandas as pd
import json
from pathlib import Path

def transfrom_json(folder_path):
    dfs = []

    for chat_id, path in enumerate(sorted(Path(folder_path).glob("*.json")), start=1):
        with open(path, encoding="utf-8") as f:
            chat_data = json.load(f)
        if "messages" in chat_data:
            chat_data = chat_data["messages"]
        df = pd.DataFrame({
            "chat_id": chat_id,
            "chat_name" : path.stem,
            "turn":    range(1, len(chat_data) + 1),
            "role":    [m["role"]    for m in chat_data],
            "content": [m["content"] for m in chat_data],
        })
        dfs.append(df)

    #hier alternativ für verschieden datei format
    if not dfs:
        raise FileNotFoundError(f"Keine JSON-Dateien in '{folder_path}' gefunden.")

    return pd.concat(dfs, ignore_index=True)

df = transfrom_json(json_path)

```
